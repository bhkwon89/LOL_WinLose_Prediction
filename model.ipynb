```python
import pandas as pd
import numpy as np



df = pd.read_csv("final_match_info2.csv")

for i in range(3, 6):
    df_tmp = pd.read_csv(f"final_match_info{i}.csv")
    df = pd.concat([df, df_tmp])
    

df = df.drop(['Unnamed: 0'], axis=1)
df = df.drop_duplicates()
df = df.reset_index(drop=True)



df['blue_top_winRate'] = ''
df['blue_top_pickRate']= ''
df['blue_top_banRate']= ''
df['blue_top_laneRate']= ''


df['blue_jg_winRate']= ''
df['blue_jg_pickRate']= ''
df['blue_jg_banRate']= ''
df['blue_jg_laneRate']= ''

df['blue_mid_winRate']= ''
df['blue_mid_pickRate']= ''
df['blue_mid_banRate']= ''
df['blue_mid_laneRate']= ''

df['blue_ad_winRate']= ''
df['blue_ad_pickRate']= ''
df['blue_ad_banRate']= ''
df['blue_ad_laneRate']= ''

df['blue_sup_winRate']= ''
df['blue_sup_pickRate']= ''
df['blue_sup_banRate']= ''
df['blue_sup_laneRate']= ''


df['red_top_winRate']= ''
df['red_top_pickRate']= ''
df['red_top_banRate']= ''
df['red_top_laneRate']= ''


df['red_jg_winRate']= ''
df['red_jg_pickRate']= ''
df['red_jg_banRate']= ''
df['red_jg_laneRate']= ''

df['red_mid_winRate']= ''
df['red_mid_pickRate']= ''
df['red_mid_banRate']= ''
df['red_mid_laneRate']= ''

df['red_ad_winRate']= ''
df['red_ad_pickRate']= ''
df['red_ad_banRate']= ''
df['red_ad_laneRate']= ''

df['red_sup_winRate']= ''
df['red_sup_pickRate']= ''
df['red_sup_banRate']= ''
df['red_sup_laneRate']= ''

df

```

```python
df_champId = pd.read_csv("champ_info.csv")



```

```python


for i in range(len(df)):
    for color in ['blue', 'red']:
        for lane in ['top', 'jg', 'mid', 'ad', 'sup']:

            champ = color + '_' + lane + '_' + "champ"
            champId = df[champ][i].split("/")[-1].split(".")[0]
            
            spell1 = color + '_' + lane + '_' + "sp1"
            spell2 = color + '_' + lane + '_' + "sp2"
            
            
            df.loc[i, spell1] = df[spell1][i].split("/")[-1].split(".")[0]
            df.loc[i, spell2] = df[spell2][i].split("/")[-1].split(".")[0]
            
            #print(champId)
            #print(df_champId[df_champId['champId'] == int(164)]['winRate'].values[0])
            
            df.loc[i, color + '_' + lane + '_' + "winRate"] = float(df_champId[df_champId['champId'] == int(champId)]['winRate'].values[0].replace("%", ""))
            df.loc[i, color + '_' + lane + '_' + "pickRate"] = float(df_champId[df_champId['champId'] == int(champId)]['pickRate'].values[0].replace("%", ""))
            df.loc[i, color + '_' + lane + '_' + "banRate"] = float(df_champId[df_champId['champId'] == int(champId)]['banRate'].values[0].replace("%", ""))
            df.loc[i, color + '_' + lane + '_' + "laneRate"] = float(df_champId[df_champId['champId'] == int(champId)][lane+'Rate'].values[0].replace("%", ""))
            
            df.loc[i, champ] = df_champId[df_champId['champId'] == int(champId)]['champName'].values[0]

df 
```

```python
def makeBinary(x):
    if x == '승리팀':
        return 0
    elif x == '패배팀':
        return 1
        
    
df['blue_winlose'] = df['blue_winlose'].apply(makeBinary)
df['red_winlose'] = df['red_winlose'].apply(makeBinary)
```

```python
tmp_col = []
for col in df.columns:
    if 'sp' in col:
        
        tmp_col += [col]
        
tmp_col
```

```python
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV

from category_encoders import TargetEncoder
from sklearn.impute import SimpleImputer 

categorical_features=['blue_top_tier', 'blue_top_champ', 'red_top_tier', 'red_top_champ',
       'blue_jg_tier', 'blue_jg_champ', 'red_jg_tier', 'red_jg_champ',
       'blue_mid_tier', 'blue_mid_champ', 'red_mid_tier', 'red_mid_champ',
       'blue_ad_tier', 'blue_ad_champ', 'red_ad_tier', 'red_ad_champ',
       'blue_sup_tier', 'blue_sup_champ', 'red_sup_tier', 'red_sup_champ',]

pipe = make_pipeline(
    
    TargetEncoder(),
    
    LGBMClassifier()
    #CatBoostClassifier(cat_features=categorical_features)
)


def divide100(x):
    return x/100

dists = {
    
    'targetencoder__smoothing': [50],# int로 넣으면 error(bug) 
    'targetencoder__min_samples_leaf': [10],
    
    'lgbmclassifier__max_depth'         :[350],
    'lgbmclassifier__learning_rate' : [0.01],
    'lgbmclassifier__n_estimators': [400]
    
    
}


pipe.get_params()
```

```python

clf = GridSearchCV(
    pipe,
    param_grid  = dists,
    #n_iter = 10,
    cv = 5,
    scoring = 'f1',
    verbose = 10,
    n_jobs = 1#, error_score='raise'
)

clf.fit(X_train, y_train)

```

```python
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier


from category_encoders import TargetEncoder
from sklearn.impute import SimpleImputer 

categorical_features=['blue_top_tier', 'blue_top_champ', 'red_top_tier', 'red_top_champ',
       'blue_jg_tier', 'blue_jg_champ', 'red_jg_tier', 'red_jg_champ',
       'blue_mid_tier', 'blue_mid_champ', 'red_mid_tier', 'red_mid_champ',
       'blue_ad_tier', 'blue_ad_champ', 'red_ad_tier', 'red_ad_champ',
       'blue_sup_tier', 'blue_sup_champ', 'red_sup_tier', 'red_sup_champ',]

pipe = make_pipeline(
    
    TargetEncoder(), 
    SimpleImputer(), 
    RandomForestClassifier(random_state=2)
    #CatBoostClassifier(cat_features=categorical_features)
)


def divide100(x):
    return x/100

dists = {
    
    'targetencoder__smoothing': [50],# int로 넣으면 error(bug) 
    'targetencoder__min_samples_leaf': [10],
    
    'randomforestclassifier__n_estimators':[450] ,
    'randomforestclassifier__max_depth': [ 30], 
    #'randomforestclassifier__max_features': (0, 1)
    
    
}


pipe.get_params()
```

```python

clf = GridSearchCV(
    pipe,
    param_grid  = dists,
    #n_iter = 10,
    cv = 5,
    scoring = 'f1',
    verbose = 10,
    n_jobs = 1#, error_score='raise'
)

clf.fit(X_train, y_train)
```
